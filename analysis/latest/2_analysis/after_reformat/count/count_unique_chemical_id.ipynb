```python
chemical_id_count = 0
for chemical_id in np.unique(morphological_data['chemical.id']):
    chemical_id_count += 1
print ("chemical_id_count:"+str(chemical_id_count))

np.savetxt("unique_chemical_id.csv", np.unique(morphological_data['chemical.id']), delimiter=",", fmt='%f')
```

```python
chemical_id_plate_id_count = 0
for chemical_id in np.unique(morphological_data['chemical.id']):
    located = morphological_data.loc[morphological_data['chemical.id'] == chemical_id,['chemical.id', 'plate.id']]
    for plate_id in np.unique(located['plate.id']):
        chemical_id_plate_id_count += 1
print ("chemical_id_plate_id_count:"+str(chemical_id_plate_id_count))
```
